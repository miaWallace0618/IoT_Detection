## 2. Load Modified DHCP files
* read DHCP files into dataframe
* modify the timestamp into right format and data type

```python
class MultiLogToDataFrame(pd.DataFrame):
    """MultiLogToDataFrame: Converts multiple Bro logs to a Pandas DataFrame
        Args:
            log_fllename (string): The full path to the Bro log
            ts_index (bool): Set the index to the 'ts' field (default = True)
        Notes:
            This class is fairly simple right now but will probably have additional
            functionality for formal type specifications and performance enhancements
    """
    def __init__(self, log_filename, ts_index=False):
        """Initialize the LogToDataFrame class"""

        # Create a bro reader on a given log file
        reader = bro_multi_log_reader.BroMultiLogReader(log_filename)

        # Create a Pandas dataframe from reader
        super(MultiLogToDataFrame, self).__init__(reader.readrows())

        # Set the index
        if ts_index:
            self.set_index('ts', inplace=True)
```

## 4. Making June_DHCP Table for Ric

```python
dhcp.head()
```

```python
dhcp.drop(dhcp.columns[[1, 2, 3,4,5,7,9]], axis=1, inplace = True)
dhcp.head()
```

```python
June_dhcp = pd.DataFrame(columns = \
                         ['min_date', 'min_ts','min_time','max_date','max_ts',\
                          'max_time','mac','assigned_ip'])

for i in range(0, all_ip.shape[0]):
    test0 = dhcp[dhcp['assigned_ip'].apply(lambda x: True if x == all_ip[i] else False)]
    test0_allmac = test0.groupby('mac').assigned_ip.count().index.values
    
    for m in range(0,test0_allmac.shape[0]):
        test0_0 = test0[test0['mac'].apply(lambda x: True if x is test0_allmac[m] else False)]
        test0_0_row = [test0_0.new_date.min(), test0_0.ts.min().strftime('%M:%S.%f')[:-4], \
                test0_0.ts.min().strftime('%M:%S.%f')[:-4], test0_0.new_date.max(), \
                test0_0.ts.max().strftime('%M:%S.%f')[:-4],test0_0.ts.max().strftime('%M:%S.%f')[:-4],\
                test0_0['mac'].iloc[0], test0_0['assigned_ip'].iloc[0]]
        June_dhcp = June_dhcp.append(pd.Series(test0_0_row, index = \
                         ['min_date', 'min_ts','min_time','max_date','max_ts',\
                          'max_time','mac','assigned_ip']),ignore_index=True)
June_dhcp.head()
```
